FUNCTION CALLING

```python
def add():
  sum=a+b
  print(sum)
a=5
b=6
add()
```

```python
def add(a,b,sum):
  print("a=",a)
  print("b=",b)
  print("a+b=",sum)
add(5,4,a+b)
```

ARGS

```python
def add(*args):
  sum=0
  for i in args:
    sum+=i
  return sum
print(add(2,3,4))
```

KWARGS

```python
def greets(**kwargs):
  print("WELCOME",end=" ")
  for key,value in kwargs.items():
    print(value,end=" ")
greets(NAME="MAYUR",USN="01FE22BEC438",BRANCH="E&C")
```
