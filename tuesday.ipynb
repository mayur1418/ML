WHILE LOOP

USING WHILE LOOP CHECK WHETHER ELIGIBLE FOR VOTING OR NOT

```python
age=int(input("ENTER THE AGE:"))
while age>=18:
  print("ELIGIBLE")
  break
else:
    print("NOT ELIGIBLE")
```
